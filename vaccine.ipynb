```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from datetime import datetime as dt
import time
```

```python
def backToSearch():
    success = False
    while success == False:
        refreshSuccess = refreshSearchResults()
        if refreshSuccess != True:
            browser.get('https://vaccine.heb.com/scheduler')
            WebDriverWait(browser, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.sc-jSgupP')))
        searchZipCode()
        WebDriverWait(browser, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.sc-fodVxV')))
        success = chooseLocation()
```

```python
def searchZipCode():
    try:
        zipCodeBox = browser.find_element_by_name('search-input')
    #     zipCodeBox = browser.find_element_by_xpath('/html/body/div/div/div[3]/div[2]/div/div/input')
        zipCodeBox.clear();
        zipCodeBox.send_keys('78758')
        searchButton = browser.find_element_by_xpath('/html/body/div/div/div[3]/div[2]/div/div/button')
        browser.execute_script("arguments[0].scrollIntoView();", searchButton)
        searchButton.click()
    except:
        print('search zip code error')
```

```python
def refreshSearchResults():
    try:
        refreshButton = browser.find_element_by_xpath('/html/body/div/div/div[3]/div[2]/div/button')
        browser.execute_script("arguments[0].scrollIntoView();", refreshButton)
        refreshButton.click()
        return True
    except:
        return False
```

```python
def chooseLocation():
    results = browser.find_element_by_css_selector('.sc-fodVxV')
    try:
#         time.sleep(1)
        items = results.find_element_by_xpath('/html/body/div/div/div[3]/div[2]/ul')
        item = items.find_elements_by_tag_name('li')[0]
    except:
        print('No open appointments')
        return False
    try:
        dist = float(item.find_elements_by_tag_name("p")[-1].text.split(" ",1)[0])
    except:
#         print("Couldn't find distance")
        return False
    address = item.find_elements_by_tag_name("address")[0].text
    if dist <= 30:
        try:
            print('Trying ' + address)
            item.find_element_by_link_text('View times').click()
            time.sleep(2)
            if ('Appointments are no longer available for this location.' in browser.page_source):
                print('Appointments are no longer available for this location.')
                return False
            if ('Appointment Details' in browser.page_source):
                if ('There are no available time slots.' in browser.page_source):
                    print('There are no available time slots.')
                    return False
                else:
                    test = chooseAppointment()
                    if test == True:
                        print('Ending Process')
                        return True
                    else:
                        print('Retrying...')
                        return False
        except:
#             print("No appointments")
            return False
    else:
#         print('No appointments within 30 miles')    
        return False
```

```python
earliestDate = dt.strptime('3/29/2021','%m/%d/%Y')
```

```python
def chooseAppointment():
    WebDriverWait(browser, 50).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'lightning-combobox.slds-form-element:nth-child(2) > div:nth-child(2) > lightning-base-combobox:nth-child(1)')))

    
    dateComboBox = browser.find_element_by_css_selector('lightning-combobox.slds-form-element:nth-child(2) > div:nth-child(2) > lightning-base-combobox:nth-child(1)')
    timeComboBox = browser.find_element_by_css_selector('lightning-combobox.slds-form-element:nth-child(3) > div:nth-child(2) > lightning-base-combobox:nth-child(1)')

    dateComboBox.click()
    dateOptions = dateComboBox.find_elements_by_tag_name('lightning-base-combobox-item')
    for dateOption in dateOptions:
        date = dt.strptime(dateOption.get_attribute('data-value'),'%m/%d/%Y')
        if date <= earliestDate:
            continue
        else:
            dateOption.click()
            timeComboBox.click()
            timeOptions = timeComboBox.find_elements_by_tag_name('lightning-base-combobox-item')[-1].click()


            #continue button
            WebDriverWait(browser, 50).until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.slds-button')))

            continueButton = browser.find_element_by_css_selector('.slds-button')
            browser.execute_script("arguments[0].scrollIntoView();", continueButton)
            continueButton.click()

            time.sleep(2)
            if ('This timeslot is full' in browser.page_source):
                print('This timeslot is full')
                return False
            else:
                print("Success!")
                return True
    print('No dates 3/29 or later')
    return False
```
